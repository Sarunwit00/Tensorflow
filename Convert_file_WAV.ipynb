### อัปโหลดไฟล์เสียง

ใช้ปุ่มด้านล่างเพื่ออัปโหลดไฟล์เสียงที่คุณต้องการแปลง (เช่น `.mp3`, `.m4a`, `.ogg`) ไฟล์ที่อัปโหลดจะถูกย้ายไปยังโฟลเดอร์ `input_audio` โดยอัตโนมัติ

```python
import os
from pydub import AudioSegment

# กำหนดโฟลเดอร์ต้นทางและปลายทาง
input_directory = 'input_audio'
output_directory = 'output_wav'

# สร้างโฟลเดอร์ปลายทางหากยังไม่มี
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# นามสกุลไฟล์ที่รองรับ
supported_extensions = ('.mp3', '.m4a', '.ogg', '.flac', '.aac', '.wma', '.aiff')

print(f"--- เริ่มต้นการแปลงไฟล์จากโฟลเดอร์: {input_directory} ---")

count = 0
# วนลูปจัดการทุกไฟล์ในโฟลเดอร์
for filename in os.listdir(input_directory):
    if filename.lower().endswith(supported_extensions):
        file_path = os.path.join(input_directory, filename)
        file_name_without_ext = os.path.splitext(filename)[0]
        output_file_path = os.path.join(output_directory, f"{file_name_without_ext}.wav")

        try:
            # โหลดและแปลงไฟล์
            audio = AudioSegment.from_file(file_path)
            audio.export(output_file_path, format="wav")
            print(f"✅ สำเร็จ: {filename} -> {output_file_path}")
            count += 1
        except Exception as e:
            print(f"❌ ข้อผิดพลาดในไฟล์ {filename}: {e}")
    else:
        if not os.path.isdir(os.path.join(input_directory, filename)):
            print(f"⚠️ ข้ามไฟล์: {filename} (นามสกุลไม่รองรับ)")

print(f"\n--- เสร็จสมบูรณ์! แปลงไฟล์ทั้งหมด {count} ไฟล์ ---")
print(f"คุณสามารถดาวน์โหลดไฟล์ได้จากโฟลเดอร์: {output_directory}")
```

### วิธีใช้งาน:

1.  **สร้างโฟลเดอร์ `input_audio`:** ใน Colab Files (คลิกที่ไอคอนโฟลเดอร์ทางซ้ายมือ) ให้สร้างโฟลเดอร์ชื่อ `input_audio`
2.  **อัปโหลดไฟล์เสียงของคุณ:** อัปโหลดไฟล์เสียงที่คุณต้องการแปลง (เช่น `.mp3`, `.m4a`) ไปยังโฟลเดอร์ `input_audio` ที่สร้างขึ้น
3.  **รันโค้ด:** รันเซลล์โค้ดด้านบน
4.  **ตรวจสอบผลลัพธ์:** ไฟล์ `.WAV` ที่แปลงแล้วจะถูกบันทึกในโฟลเดอร์ `output_wav` ซึ่งจะถูกสร้างขึ้นอัตโนมัติ

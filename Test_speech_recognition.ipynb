```python
import speech_recognition as sr
from google.colab import files
from pydub import AudioSegment
import os

# 1. อัปโหลดไฟล์เสียง (.m4a)
print("กรุณาอัปโหลดไฟล์เสียง (.m4a)...")
uploaded = files.upload()

if len(uploaded) > 0:
    # ดึงชื่อไฟล์ที่อัปโหลดเข้ามา
    original_file_name = list(uploaded.keys())[0]
    print(f"\nอัปโหลดไฟล์ '{original_file_name}' สำเร็จ")

    # 2. แปลงไฟล์จาก .m4a เป็น .wav
    print("กำลังแปลงไฟล์เป็นรูปแบบที่รองรับ (.wav)...")
    wav_file_name = "converted_audio.wav"

    try:
        # อ่านไฟล์ .m4a และแปลง (Export) เป็น .wav
        audio = AudioSegment.from_file(original_file_name, format="m4a")
        audio.export(wav_file_name, format="wav")
        print("แปลงไฟล์สำเร็จ! กำลังถอดรหัสเสียงเป็นข้อความ...\n")

        # 3. เริ่มกระบวนการแปลงเสียงเป็นข้อความ
        r = sr.Recognizer()
        with sr.AudioFile(wav_file_name) as source:
            audio_data = r.record(source)

            try:
                # ส่งไปให้ Google ประมวลผล (ภาษาไทย)
                text = r.recognize_google(audio_data, language="th-TH")

                print("--- ข้อความที่ได้จากเสียง ---")
                print(text)
                print("---------------------------")

            except sr.UnknownValueError:
                print("ข้อผิดพลาด: Google ไม่สามารถเข้าใจเสียงนี้ได้ (เสียงอาจไม่ชัดเจน)")
            except sr.RequestError as e:
                print(f"ข้อผิดพลาดจากเซิร์ฟเวอร์: {e}")

    except Exception as e:
        print(f"เกิดข้อผิดพลาดในการแปลงไฟล์เสียง: {e}")

    finally:
        # 4. ลบไฟล์ที่แปลงชั่วคราวทิ้งเพื่อประหยัดพื้นที่ (ทางเลือก)
        if os.path.exists(wav_file_name):
            os.remove(wav_file_name)
else:
    print("ไม่มีการอัปโหลดไฟล์")
```
